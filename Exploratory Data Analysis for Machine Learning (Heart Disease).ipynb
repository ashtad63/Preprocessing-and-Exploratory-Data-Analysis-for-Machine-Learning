### Step B

```python
heart_corr = heart_data.corr()
grps = ['NegStrong','NegMedium','Weak', 'PosMedium', 'PosStrong']
cols = ['Age', 'Sex', 'Pain', 'BP', 'Chol', 'BS', 'Cardio', 'HR', 'Angina', 'STPeak','STSlope', 'NumVess', 'Thal', 'HDis']

for col in cols:
    heart_corr[col] = pd.cut(heart_corr[col], (-1,-0.8,-0.6,0.6,0.8,1), labels=grps)
heart_corr
```

### Step C

```python
pain_dum = pd.get_dummies (heart_data['Pain'], prefix = 'Pain_')
card_dum = pd.get_dummies (heart_data['Cardio'], prefix = 'Card_')
thal_dum = pd.get_dummies (heart_data['Thal'], prefix = 'Thal_')
```

```python
heart_X1 = heart_data[['Age', 'Sex']].join(pain_dum)
heart_X2 = heart_data[['BP', 'Chol', 'BS']].join(card_dum)
heart_X3 = heart_data[['HR', 'Angina', 'STPeak', 'STSlope', 'NumVess']].join(thal_dum)
heart_Y = heart_data[['HDis']]
```

```python
heart_data = pd.concat([heart_Y, heart_X1, heart_X2, heart_X3], axis=1)
heart_data.head(20)
```
